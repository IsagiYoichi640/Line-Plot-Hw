```python
import matplotlib.pyplot as plt
```

```python

days=["Mon", "Tue", "Wed","Thu","Fri","Sat","Sun"]
players_active =[10000,7000,5000,2000,3000,15000,6000]
plt.plot(days, players_active, marker='o', linestyle='-',color='r')
plt.title("Players Active in My Roblox Game Over a Week")
plt.xlabel('Days')
plt.ylabel('Players_Active')
plt.show()
```

# Conclusion
* The highest amount of player active was in Saturday with 15000 player playing my game
* The lowest amount of player active was in Thursday with 2000 player playing my game
* The number of active players varies significantly throughout the week.
* There is a noticeable spike in players during Saturday, followed by a decline on Sunday, which may mean players are more active on Saturdays but slightly less on Sundays.
